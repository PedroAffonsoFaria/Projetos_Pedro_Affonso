# Gerenciamento dos Insumos

```python
# Lista de produtos Ãºnicos
produtos_unicos = data['produto'].unique()

# DicionÃ¡rio para armazenar previsÃµes por produto
previsoes_por_produto = {}

# Treinar um modelo para cada produto
for produto in produtos_unicos:
    # Filtrar os dados para o produto especÃ­fico
    vendas_produto = data[data['produto'] == produto]

    # VariÃ¡veis independentes e dependentes
    X = vendas_produto[['semana']]
    y = vendas_produto['quantidade']

    # Handle missing values in 'semana' column, if any
    X = X.dropna()  # Drop rows with missing 'semana' values
    y = y[X.index]  # Align y with the dropped rows in X

    # Modelo de RegressÃ£o Linear
    modelo = LinearRegression()
    modelo.fit(X, y)

    # Prevendo para as prÃ³ximas semanas (exemplo: semanas 24, 25, 26)
    semanas_futuras = pd.DataFrame({'semana': [23,24, 25, 26]})
    previsoes = modelo.predict(semanas_futuras)

    # Armazenar as previsÃµes
    previsoes_por_produto[produto] = previsoes

# Converte as previsÃµes para um DataFrame
previsoes_df = pd.DataFrame(previsoes_por_produto)
previsoes_df['semana'] = semanas_futuras['semana']

# Mostrar as previsÃµes por produto
previsoes_df
```

```python
# Combina os dados reais e previstos
produto_especifico = 'PANDORA 500ML'
vendas_reais = data[data['produto'] == produto_especifico][['semana', 'quantidade']]
vendas_reais.columns = ['semana', 'real']
vendas_previstas = previsoes_df[['semana', produto_especifico]]
vendas_previstas.columns = ['semana', 'previsto']

# Concatenar os dados reais e previstos, including all columns of vendas_previstas
vendas_combinadas = pd.concat([vendas_reais, vendas_previstas], axis=0).reset_index(drop=True) # Changed this line

vendas_previstas
```

```python
# Calcular o consumo total previsto para cada ingrediente
receitas = {
    'PANDORA 500ML': {'Nutella': 30, 'Brigadeiro de Leite Ninho': 30, 'Leite Ninho': 3},
    'PANDORA 300ML': {'Nutella': 20, 'Brigadeiro de Leite Ninho': 20, 'Leite Ninho': 2.5},
    'PRECIOSA 300ML': {'Nutella': 40, 'Ferreiro Roche': 1},
    'MIMOSA 300ML': {'Brigadeiro de Leite Ninho': 40, 'Calda de Morango': 1.33, 'Leite Ninho': 1},
    'MALHADA 300ML': {'Brigadeiro de Leite Ninho': 40, 'Oreo': 3, 'Leite Ninho': 1}
}

conversoes = {
    'Nutella': 650,
    'Leite Ninho': 25,
    'Oreo': 48,
    'Calda de Morango': 4500
}

# Inicializar um dicionÃ¡rio para armazenar o consumo total de cada ingrediente
consumo_previsto = {
    'Nutella': 0,
    'Brigadeiro de Leite Ninho': 0,
    'Leite Ninho': 0,
    'Oreo': 0,
    'Calda de Morango': 0,
    'Ferreiro Roche': 0
}
```

```python
# Iterar pelas previsÃµes
for produto, previsoes in previsoes_por_produto.items():
    for quantidade in previsoes:
        # Verificar os ingredientes para cada produto
        if produto in receitas:
            for ingrediente, quant in receitas[produto].items():
                consumo_previsto[ingrediente] += quant * quantidade
```

```python
# Converter o consumo para unidades de estoque (ex: potes, caixas)
consumo_em_unidades = {ingrediente: consumo / conversoes.get(ingrediente, 1)
                       for ingrediente, consumo in consumo_previsto.items()}

# Exibir o consumo previsto em unidades de estoque
for ingrediente, consumo in consumo_em_unidades.items():
    print(f"{ingrediente}: {consumo:.2f} unidades")
```
